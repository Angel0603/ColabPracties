```python
import pandas as pd
import random
df = pd.read_csv('Iris.csv')
```

```python
indices = list(df.index)
random.shuffle(indices)
df_revuelto = df.iloc[indices]
df_revuelto.head(15)
```

```python
x_train = df_revuelto.iloc[:120,1:5]
x_train
```

```python
y_train = df_revuelto.iloc[:120,-1]
y_train
```

```python
x_test = df_revuelto.iloc[120:,1:5]
x_test
```

```python
y_test = df_revuelto.iloc[120:,-1]
y_test
```

```python
from sklearn.svm import SVC
modelo = SVC()
modelo.fit(x_train,y_train)
```

```python
y_predic = modelo.predict(x_test)
y_predic
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
print("Muestra el porcentaje de precisiÃ³n en la predicciÃ³n:")
accuracy_score(y_predic,y_test)
```

```python
print("Muestra la cantidad acertada de la predicciÃ³n:")
accuracy_score(y_predic,y_test, normalize=False)
```
