```python
URL = 'https://raw.githubusercontent.com/juarezefren/datasets/main/insects-10.csv'
import pandas as pd
df = pd.read_csv(URL)
df.shape
```

```python
df
```

```python
x_train = df.iloc[:,1:3]
x_train
```

```python
y_train = df.iloc[:,-1]
y_train
```

```python
from sklearn.svm import SVC
modelo = SVC()
modelo.fit(x_train,y_train)
```

```python
x_test = pd.DataFrame({'abdomen':[5.1], 'antena':[7.0]})
x_test
```

```python
modelo.predict(x_test)
```
